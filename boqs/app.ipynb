```python
import ezdxf
```

```python
# Load the DXF file
doc = ezdxf.readfile("data/autocad.dxf")
```

```python

# Access modelspace
msp = doc.modelspace()

# Iterate through elements and get their dimensions
for entity in msp:
    print (entity)
    if entity.dxftype() == "LINE":
        print(f"Line start: {entity.dxf.start}, end: {entity.dxf.end}")
    elif entity.dxftype() == "CIRCLE":
        print(f"Circle center: {entity.dxf.center}, radius: {entity.dxf.radius}")
```

```python
import ezdxf

doc = ezdxf.readfile("data/autocad.dxf")
msp = doc.modelspace()

for entity in msp:
    if entity.dxftype() == "LINE":  # Assuming beams are lines
        start = entity.dxf.start
        end = entity.dxf.end
        length = ((end[0] - start[0])**2 + (end[1] - start[1])**2 + (end[2] - start[2])**2) ** 0.5
        print(f"Beam Length: {length}")
```
